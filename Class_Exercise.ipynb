```python
# Import the data Building_Permits
df = pd.read_csv("/content/Building_Permits.csv")
```

```python
# Print the head and tail of the data
df.head()
```

```python
# show the number of missing data points per column
df.isnull().sum()
```

```python
# show the number of missing points in the first ten columns
missing = df.iloc[:, :10].isnull().sum()
print(missing)
```

```python
# how many total missing values do we have?
y = df.isnull().sum().sum()
print(y)
```

```python
# percent of data that is missing
missing = (y / df.size) * 100
print(missing)
```

```python
# remove all the rows that contain a missing value
df.dropna(axis = 0)
```

```python
# remove all columns with at least one missing value
df.dropna(axis = 1)
```

```python
# calculate how much data did we lose? in percentage
lose = df.isnull().sum().sum()
lose_percen = ((lose / df.size)* 100)
print(lose_percen)
```

```python
# get a small subset of the Building_ permits dataset
df[["Permit Number", 'Permit Type', "Permit Type Definition", "Permit Creation Date"]]
```

```python
# replace all NA's with 0
df.fillna(0)
```

```python
# replace all NA's with the value that comes directly after it in the same column, then replace all the reamining na's with 0
new_building = df.fillna(method='bfill')
new_building.fillna(0,inplace=True)
```
